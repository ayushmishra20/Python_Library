### SEABORN : HEATMAP


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.head()
```

```python
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)

xlabels = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri']
ylabels = ['Coffee', 'Juice', 'Soda', 'Tea']

orders = np.array([[44, 31, 50, 41, 30], 
                   [21, 12, 28, 38, 35],
                   [15, 18, 20, 14, 12],
                   [38, 29, 39, 26, 11]]
                 )
```

```python
sns.heatmap(orders, cmap = 'YlOrBr', vmin = 0,
           annot = True,
           alpha = 0.8,
           square = True,
           annot_kws={'fontsize':14, 'fontweight': 'bold', 'color': 'black'})
```

```python
test_array = np.array([[1,2,3], [4,5,6]])
test_array
```

```python
sns.heatmap(test_array, annot = True)
```

```python
Transpose_Mat = test_array.T
```

```python
sns.heatmap(Transpose_Mat, annot = True)
```

```python
cars = sns.load_dataset('mpg')
cars.groupby('origin').cylinders.value_counts()
```

```python
origin_cyl = (cars
 .groupby('origin')
 .cylinders
 .value_counts()
 .unstack()
 .fillna(0)
)
```

```python
sns.heatmap(origin_cyl)
plt.tight_layout()
plt.show()
```

```python
sns.heatmap(origin_cyl.T, cmap = 'RdBu')
plt.tight_layout()
```

```python
labels = ['MPG', 'Cylinders', 'Displacement', 'Horsepower',
          'Weight', 'Acceleration', 'Model Year'
         ]
```

```python
sns.heatmap(cars.corr(), cmap = 'Blues', annot = True, linewidth = 1, linecolor = 'black')
plt.tight_layout()
```

```python
sns.heatmap(cars.corr(), 
            cmap='RdBu',
            xticklabels=labels,
            yticklabels=labels,
            annot = True
           );
```

```python
sns.heatmap(cars.corr(), 
            cmap='RdBu', 
            xticklabels=False,
            yticklabels=labels
           );
```

```python
sns.heatmap(cars.corr(), 
            cmap='RdBu',
            square=True
           );
```
