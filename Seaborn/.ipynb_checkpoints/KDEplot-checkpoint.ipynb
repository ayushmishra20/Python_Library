### SEABORN : KDEplot

#### UNIVARIAT(1D)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
cars.head()
```

```python
sns.set_style('darkgrid')
sns.kdeplot(cars.horsepower, fill = True, bw_method = 5)
plt.show()
cars.shape
```

```python
sns.kdeplot(cars.horsepower, fill = True, cumulative = True, legend = True )
plt.show()
```

#### Bivariate(2d)

```python
sns.kdeplot(x = 'horsepower', y = 'mpg', data = cars, alpha = 0.4)
sns.scatterplot(x = 'horsepower', y = 'mpg', data = cars, color = 'black', alpha = 0.5)
plt.xlabel('HorsePower', fontsize = 14)
plt.ylabel('MPG', fontsize = 14)
plt.tight_layout()
plt.show()
```

```python
sns.kdeplot(x = 'horsepower', y = 'mpg', data = cars)
plt.xlim(0,260)
plt.ylim(0,55)
plt.xlabel('HorsePower', fontsize = 14)
plt.ylabel('MPG', fontsize = 14)
plt.show()
```

```python
sns.kdeplot(x = 'horsepower', y = 'mpg', data = cars, n_levels = 20, fill = True, thresh = False,
           cbar = True)
plt.xlabel('Horsepower', fontsize = 14)
plt.ylabel('MPG', fontsize = 14)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.distplot(cars.horsepower, kde_kws={"lw": 5}, hist_kws={'alpha': 0.25})

sns.despine(left=True)
plt.xlim(0, 280)
plt.xlabel('')
plt.yticks([])
plt.tight_layout();
```
