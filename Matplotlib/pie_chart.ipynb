```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('ggplot')
# pie chart shoud use when the item is <5
# neither use bar chart 
```

```python
slices = [60,50,23,56]
labels = ['sixty', 'fifty', 'twentythree', 'fiftysix']
plt.pie(slices, labels = labels, wedgeprops={'edgecolor':'black'})
plt.title('Pie chart')
plt.tight_layout()
plt.grid()
plt.show()
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]
plt.pie(slices, labels = labels, explode = explode, shadow = True,startangle = 90,autopct = '%1.1f%%' , wedgeprops = {'edgecolor':'black'})
plt.title('Pie Chart for Language popularity')

plt.show()


```
