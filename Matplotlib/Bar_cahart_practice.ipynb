```python
import csv
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np 
import pandas as pd
```

```python
plt.style.use('ggplot')
data= pd.read_csv("C:\\Users\\Ayush mishra\\Downloads\\data.csv")
ids = data['Responder_id']
lang_resonses = data['LanguagesWorkedWith']

language_counter = Counter()


for response in lang_resonses:
    language_counter.update(response.split(';'))

# print(language_counter.most_common(15)) # gives a list of common 
language = []
popularity = []

for item in language_counter.most_common(15):
    language.append(item[0])
    popularity.append(item[1])

print(language)
print(popularity)
language.reverse()
popularity.reverse()
plt.barh(language, popularity, color= 'blue')
plt.title('Language Popularity')
# plt.ylabel('Language')

plt.tight_layout()
plt.xlabel('Popularity')
plt.legend()
plt.show()
```
