```python
import pandas as pd

data = {
    "Name": ["Aman", "Ravi", "Aman", None],
    "Marks": [85, None, 90, 75],
    "City": ["Delhi", "Mumbai", "Delhi", "Delhi"]
}

df = pd.DataFrame(data)
df
```

```python
df.isnull()
```

```python
df.notnull()
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.fillna("unknown")
```

```python
df["Marks"].fillna(df["Marks"].mean())
```

```python
df.duplicated()
```

```python
df.duplicated(subset=["Name"])
```

```python
df.drop_duplicates()
```

```python
df.drop_duplicates(subset = ["Name"])
```

```python
df.head()
```

```python
df["Marks"] = df["Marks"].astype(float)
```

```python
df.groupby("City")["Marks"].mean()

```

```python
df.groupby("City")["Marks"].count()
```

```python
df.groupby("City").agg({
    "Marks": ["mean", "sum", "count"]
})
```
