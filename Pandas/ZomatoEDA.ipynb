```python
final_df=pd.merge(df,df_country,on ='Country Code', how='left')
```

```python
#to check data type
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_value=final_df.Country.value_counts().values
```

```python
# Pie Chart
plt.pie(country_value[:3], labels =country_names[:3], autopct='%1.2f%%' )
plt.show()
```

#Observation :

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Conclusions
1.When Rating is between 4.5 to 4.9---->Excellent
2.When Rating between 4.0 to 3.4---->very good
3.When Rating between 3.5 to 3.9---->good
4.When Rating between 3.0 to 3.4---->average
5.When Rating between 2.5 to 2.9---->Average
6.When Rating between 2.0 to 2.4---->poor


```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12, 6)
sns.barplot(x="Aggregate rating",y="Rating Count", data=ratings)
plt.show()
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color', data=ratings,palette=['Blue','red','yellow', 'green'])
plt.show()
```

#Observations
    1.Not Rated count is very high
    2.Maximum number of rating are between 2.5 to 3.4

```python
##Count Plot
sns.countplot(x="Rating color",data=ratings,palette=['green','red','yellow', 'darkgreen'])
plt.show()
```
