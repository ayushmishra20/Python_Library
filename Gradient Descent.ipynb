## In Formal Language: First-Order Optimization

Gradient descent is a first-order iterative optimization algorithm used to find a local minimum of a differentiable function. The algorithm operates by taking proportional steps in the opposite direction of the gradient (the multi-variable generalization of the derivative) of the function at the current point, as this represents the direction of steepest descent.Mathematically, if you have a cost function $J(\theta)$ that you want to minimize, the algorithm updates the parameters $\theta$ using the following rule:$$ \theta_{n+1} = \theta_n - \alpha \nabla J(\theta_n) $$Where:$\theta_{n+1}$ is the next position (the updated weights/variables).$\theta_n$ is the current position.$\alpha$ is the learning rate (a positive scalar determining the size of the step).$\nabla J(\theta_n)$ is the gradient of the cost function evaluated at the current position.If the cost function is strictly convex (shaped like a perfect bowl), the algorithm is mathematically guaranteed to converge to the global minimum. For non-convex functions with many valleys—like deep neural networks—it converges to a local minimum.
