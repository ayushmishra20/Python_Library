```python
country_respondent = df['Country'].value_counts()
country_respondent
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply( lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondent, country_uses_python], axis = 'columns', sort = False)
python_df
```

```python
python_df.rename(columns={'Country':'NumRespondent', 'LanguageWorkedWith': "NumKnowPython"}, inplace = True)
python_df
```

```python
python_df['PercentKnowPython'] = (python_df['NumKnowPython']/python_df['count'] * 100)
python_df
```

```python
python_df.sort_values(by = 'PercentKnowPython', ascending=False, inplace=True)
python_df.head(50)
```
